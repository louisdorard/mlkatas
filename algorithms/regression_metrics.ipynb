### Question

Complete the following functions to compute MAE, MSE, MAPE and R^2.

```python
def mae(y_true, y_pred):
    y_err = abs(y_pred - y_true)
    return sum(y_err)/len(y_err)
    # YOUR CODE HERE
```

```python
def mse(y_true, y_pred):
    return np.mean((y_pred - y_true)**2)
    # YOUR CODE HERE
```

```python
def r2(y_true, y_pred):
    return 1 - mse(y_true, y_pred) / mse(y_true, np.mean(y_true))
    # YOUR CODE HERE
```

```python
def mape(y_true, y_pred):
    return np.mean(abs(y_pred - y_true)/y_true)
    # YOUR CODE HERE
```
